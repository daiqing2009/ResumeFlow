```python
import json
try:
    with open("/content/eval_exp_ds.json", 'r') as json_file1:
        json1 = json.load(json_file1)
    print(len(json1))
except Exception as e:
    print(f"An error occurred: {e}")
```

```python
!pip install bert_score
```

```python
from evaluate import load
bertscore = load("bertscore")
```

```python
predictions = []
references = []
# work_experience section evaluation
for item in json1:
  predictions.append(json.dumps(item.get("inference_output"))) # fine-tuned gemma2's prediction
  references.append(json.dumps(item.get("benchmark_output"))) # gemini-1.5's reference

# results = bertscore.compute(predictions=predictions, references=references, lang="en", model_type="distilbert-base-uncased")
results = bertscore.compute(predictions=predictions, references=references, lang="en")
```

```python
print(results)
```

```python
eval_precision_results = results.get("precision")
eval_recall_results = results.get("recall")
eval_f1_results = results.get("f1")


precision_score = 0.0
for p in eval_precision_results:
  precision_score += p
overall_precision_score = precision_score/len(eval_precision_results)
# average BERTScore of work_experience section
print(f"Average Precision Score: {overall_precision_score:.3f}")


recall_score = 0.0
for p in eval_recall_results:
  recall_score += p
overall_recall_score = recall_score/len(eval_recall_results)
# average BERTScore of work_experience section
print(f"Average Recall Score: {overall_recall_score:.3f}")


f1_score = 0.0
for p in eval_f1_results:
  f1_score += p
overall_f1_score = f1_score/len(eval_f1_results)
# average BERTScore of work_experience section
print(f"Average F1 Score: {overall_f1_score:.3f}")
```
